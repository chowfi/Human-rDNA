```python
def compare_fasta_txt(fasta_file, txt_file):
    with open(txt_file, "r") as txt:
        txt_sequence = txt.read().strip().replace("\n", "")

    fasta_sequence = ""
    with open(fasta_file, "r") as fasta:
        for line in fasta:
            if not line.startswith(">"):  
                fasta_sequence += line.strip()

    if fasta_sequence == txt_sequence:
        print("✅ The FASTA file and TXT file contain the same sequence.")
    else:
        print("❌ The sequences do NOT match.")
        print(f"FASTA Length: {len(fasta_sequence)}, TXT Length: {len(txt_sequence)}")

compare_fasta_txt("1000_genome_new_ref_v3_string.fasta", "1000_genome_new_ref_v3_string.txt")
```
